```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
from huggingface_hub import login

small_embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
large_embedding_model = SentenceTransformer('deepset/sentence_bert')


query = "What is the process by which plants convert sunlight into energy?"
documents = [
    "Photosynthesis is the process by which plants convert sunlight, water, and carbon dioxide into oxygen and glucose.",
    "Cellular respiration is a process where cells produce energy from oxygen and glucose.",
    "Plants absorb water through their roots via osmosis."
]

query_embedding = small_embedding_model.encode(query, convert_to_tensor=True)
document_embeddings = small_embedding_model.encode(documents, convert_to_tensor=True)

similarity_scores = util.pytorch_cos_sim(query_embedding, document_embeddings)

top_k = 3
top_k_indices = np.argsort(similarity_scores[0].cpu().numpy())[::-1][:top_k]

print("Top-k Retrieved Documents:")
for idx in top_k_indices:
    print(f"Document: {documents[idx]}, Score: {similarity_scores[0][idx].item()}")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch


rerank_model = AutoModelForSequenceClassification.from_pretrained('cross-encoder/ms-marco-MiniLM-L-12-v2')
rerank_tokenizer = AutoTokenizer.from_pretrained('cross-encoder/ms-marco-MiniLM-L-12-v2')


def rerank(query, documents):
    rerank_inputs = [f"{query} [SEP] {doc}" for doc in documents]
    inputs = rerank_tokenizer(rerank_inputs, return_tensors="pt", padding=True, truncation=True, max_length=512)

    with torch.no_grad():
        outputs = rerank_model(**inputs)
        scores = outputs.logits[:, 0]

    return scores


top_k_docs = [documents[i] for i in top_k_indices]


rerank_scores = rerank(query, top_k_docs)


reranked_indices = torch.argsort(rerank_scores, descending=True)


print("\nReranked Documents:")
for idx in reranked_indices:
    print(f"Document: {top_k_docs[idx]}, Rerank Score: {rerank_scores[idx].item()}")
```

```python
from sklearn.metrics import ndcg_score

true_relevance = [3, 2, 1]


ndcg_value = ndcg_score([true_relevance], [rerank_scores.tolist()], k=10)
print(f"NDCG@10: {ndcg_value}")

```
